### 1. Math Symbol

- $Z^*_{n}$: it is an example of group. By definition of a group, in set $Z^*_{n}$ together with operation $*$(multiplication), that combines any two elements to form an element of this set. Operations between elements follow three requirements as below:
    - Associativity: For all $a,b,c$ in $Z^*_{n}$, one has $(a*b)*c=a*(b*c)$.
    - Identity element: There exists a unique element $e$ in $Z^*_{n}$ which is called identity element, for every $a\in Z^*_{n}$, one has $a*e=e*a$.
    - Inverse element: For every $a\in Z^*_{n}$, there exists an element $b\in Z^*_{n}$ called $a$'s inverse such that $a*b=b*a=e$. For each $a$, its inverse is unique and can be denoted as $a^{-1}$.
- $<g>$: when $Z^*_{n}$ has a generator, we call $Z^*_{n}$ a cyclic group. If $g$ is a generator we write $Z^*_{n}=<g>$.

### 2. Subgroup of $Z^*_{n}$ 

- Definition: A subgroup of $Z^*_{n}$ is a non-empty subset $H$ of $Z^*_{n}$ such that if $a,b\in H$ then $ab\in H$.

- Extension: 
    - As any $a$ has order, there exists $a^k=1$ so $a^{-1}=a^{k-1}$. Thus, every element in subset has inverse and $1$ is an element in subset.
    - Any $a\in Z^*_{n}$ can be used to generate cyclic group $<a>={a, a^2, ... a^d=1}$ for some $d$.
    - ${1}$ is always is subgroup of any group.

### 3. Lagrange's Theorem

- Theorem: Let $H$ be a subgroup of $Z^*_{n}$ of size $m$, then $m|\phi(n)$.

- Proof:
    - 1. If $H=Z^*_{n}$, then $m=\phi(n)$.
    - 2. If $H\neq Z^*_{n}$, let $H={h_1, h_2, ..., h_m}$. 
        - Let $a\in Z^*_{n}$ be some element not in $H$. Denote set $aH=\{ah_1, ah_2, ..., ah_m\}$.
        - Every element in $aH$ is distinct and $aH\bigcap H=\emptyset$
    - 2.1. If $H\bigcup aH=Z^*_{n}$, then $2m=\phi(n)$
    - 2.2. If there is $b\notin H\bigcup aH$, then we have $bH=\{bh_1, bh_2, ..., bh_m\}$ and $bH\bigcap H=\emptyset,bH\bigcap aH=\emptyset$. Then we know for any $b$ we could find a set $H'$ having $m$ elements which are not contained in sets we've found. Thus the theorem is proved.
    
- Corollary: 
    - For any $a\in Z^*_{n}$ generates a cyclic subgroup $\{a, a^2, ..., a^d=1\}$ thus $d|\phi(n)$, hence $a^{\phi(n)}=1$

### 4. Subgroups of Cyclic Groups

- Theorem: All subgroups of a cyclic group are cyclic. If $G=<g>$ is a cyclic group of order $n$ then for each divisor $d$ of $n$ there exists exactly one subgroup of order $d$ and it can be generated by $g^{n/d}$.

- Proof:
    - Existence: Given a divisor $d$, let $e=n/d$, then $\{g^e, g^{2e}, ..., g^{de}\}$ is a cyclic subgroup of $G$.
    - Uniqueness: 
        - For a cyclic subgroup of $G$ being $\{a_1, a_2, ..., a_d=1\}$, for each $a_i$ we have $a_i=g^k$ for some $k$. By $Lagrange's Theorem$, the order of $a_i$ must divide $d$, thus $g^{kd}=1$.
        - As order of $g$ is $n$, so $kd=mn=mde,\; k=me$ for any integer $m$, then we have $a_i=g^{me}=(g^e)^m$. Thus the cyclic subgroup must be unique.

### 5. Counting Generators

- Theorem1: Let $G$ be cyclic group of order $n$. Then $G$ contains exactly $\phi(n)$ generators.

- Proof1: Let $g$ be a generator of $G$, then any generator of $G$ denoting $g^k$ must satisfy $g^{km}=g$, then $km=1(mod\;n)$, so any unit in $Z_n$ is valid $k$. There are $\phi(n)$ units in $Z_n$.   

<br>

- Theorem2: For any positive integer $n$, we have 
$n=\sum_{d|n} \phi(d)$.

- Proof2: For a cyclic group $G=<g>$ whose order is $n$, each element $g^k$ is one of generators in a cyclic group. And for all cyclic groups there is no mutual generator. Thus, the amount of all generators which is $\sum_{d|n} \phi(d)$ is equal to the $n$.

### 6. Group Struture

- Group isomorphism: For every positive integer $n$, there is only one cyclic group of order $n$. 
    - Example1: $Z^*_3,\; Z^*_4$ are cyclic group of order $2$.
    - Example2: 
        - For $n=2^kp_1^{k_1}...p_m^{k_m}$ for odd primes $p_i$, by Chinese Remaider Theorem, we have 
    $Z^*_n=Z^*_{2^k}\times Z^*_{p_1^{k_1}}...\times Z^*_{p_m^{k_m}}$.
        - As $Z^*_{p_i^{k_i}},Z^*_2,Z^*_4$ are cyclic, then group structure of $Z^*_n$ is 
            - $C_{2^k-1}\times C_{p_1^{k_1}-p_1^{k_1-1}}...\times C_{p_m^{k_m}-p_m^{k_m-1}}$, when $k\in \{1,2\}$,
            - $C_2\times C_{2^k-2}\times C_{p_1^{k_1}-p_1^{k_1-1}}...\times C_{p_m^{k_m}-p_m^{k_m-1}}$, when $k\notin \{1,2\}$.
